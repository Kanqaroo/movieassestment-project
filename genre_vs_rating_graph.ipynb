```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('show_data.csv',index_col=0)
```

```python
df_network = df[df['network'] != ('False')]
```

```python
adven = df[df['Adventure']==1].groupby('Adventure')['rating'].mean()[1]
action = df[df['Action']==1].groupby('Action')['rating'].mean()[1]
comedy = df[df['Comedy']==1].groupby('Comedy')['rating'].mean()[1]
crime = df[df['Crime']==1].groupby('Crime')['rating'].mean()[1]
drama = df[df['Drama']==1].groupby('Drama')['rating'].mean()[1]
fantasy = df[df['Fantasy']==1].groupby('Fantasy')['rating'].mean()[1]
horror = df[df['Horror']==1].groupby('Horror')['rating'].mean()[1]
romance = df[df['Romance']==1].groupby('Romance')['rating'].mean()[1]
```

```python
adven1 = df[df['Adventure']==1].groupby('Adventure')['award_exposure'].mean()[1]
action1 = df[df['Action']==1].groupby('Action')['award_exposure'].mean()[1]
comedy1 = df[df['Comedy']==1].groupby('Comedy')['award_exposure'].mean()[1]
crime1 = df[df['Crime']==1].groupby('Crime')['award_exposure'].mean()[1]
drama1 = df[df['Drama']==1].groupby('Drama')['award_exposure'].mean()[1]
fantasy1 = df[df['Fantasy']==1].groupby('Fantasy')['award_exposure'].mean()[1]
horror1 = df[df['Horror']==1].groupby('Horror')['award_exposure'].mean()[1]
romance1 = df[df['Romance']==1].groupby('Romance')['award_exposure'].mean()[1]
```

```python
genre_name = ('Adven','Action','Comedy','Crime','Drama','Fantasy','Horror','Romance')
genre_rating = (adven,action,comedy,crime,drama,fantasy,horror,romance)
average_reward_rating = (adven1,action1,comedy1,crime1,drama1,fantasy1,horror1,romance1)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
barlist = ax.bar(genre_name,genre_rating, align ="center",alpha = 1)
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Rating by Genre')
plt.xticks(rotation=0)
barlist[0].set_color('lawngreen')
barlist[7].set_color('gold')
plt.ylim((8.2,8.8))
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(20, 10))
barlist1 = ax1.bar(genre_name, average_reward_rating, align='center',alpha=1)
plt.xlabel('Genre')
plt.ylabel('Average Rewards awarded')
plt.title('Rewards Awarded by Genre')
plt.xticks(rotation=0)
barlist1[7].set_color('lawngreen')
plt.show()
```

```python
type_award_list = df_network.groupby('type')['award_exposure'].mean()
```

```python
animation = type_award_list[0]
documentary = type_award_list[1]
panel_show = type_award_list[2]
reality = type_award_list[3]
script = type_award_list[4]
talk_show = type_award_list[5]
variety = type_award_list[6]
type_name = ('Animation','Documentary','Panel Show','Reality','Scripted','Talk Show','Variety')
type_award = (animation,documentary,panel_show,reality,script,talk_show,variety)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
barlist2 = ax.bar(type_name,type_award, align='center',alpha=1)
plt.xlabel('Type of TV Program')
plt.ylabel('Most Awarded TV Program')
plt.title('Average Amount of Awards')
plt.xticks(rotation=0)
barlist2[4].set_color('lawngreen')
plt.show()
```
