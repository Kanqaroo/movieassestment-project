```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
the_df['network'] = np.where(the_df['network'].str.find('BBC') >= 0, 'BBC', the_df['network'])
# takes all BBC type networks and places them under on category: 'BBC'
```

```python
the_df['webChannel'] = np.where(the_df['webChannel'].str.find('HBO Go') >= 0, 'HBO Go', the_df['webChannel'])
# takes all HBO Go type streaming servies and places them under on category: 'HBO Go'
```

```python
networks = the_df[the_df['network'] != False]
# created df which represents all shows on network TV
```

```python
streaming = the_df[the_df['webChannel'] != False]
# created df which represents all shows on a streaming platform
```

```python
simple_comp_bars = plt.bar(['Streaming Services', 'Network Services'], [streaming['rating'].dropna().mean(), networks['rating'].dropna().mean()],width=0.6,align='center', )
plt.ylabel('Average Rating')
plt.title('Avg. Ratings of Each Service')
plt.show()
```

```python
simple_comp_bars = plt.bar(['Streaming Services', 'Network Services'], [streaming['award_exposure'].dropna().mean(), networks['award_exposure'].dropna().mean()],width=0.6,align='center', )
plt.ylabel('Average Awards Received')
plt.title('Avg. Awards Exposure by Service')
plt.show()
```

```python
xscatter=networks['rating']
yscatter=networks['award_exposure']
# created x and y variable for the scatter plot
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.set_style('whitegrid')
sns.set_context('notebook')
ax = sns.regplot(x=xscatter, y=yscatter, data=networks, line_kws={'color':'red'})
plt.title('Award Exposure and Avg. Rating per Network TV Show')
plt.xlabel('Avg. Rating')
plt.ylabel('Award Exposure')
```
