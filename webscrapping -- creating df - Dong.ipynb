```python
import requests
import json
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from bs4 import BeautifulSoup
page = requests.get('https://www.imdb.com/chart/toptv/?sort=rk,asc&mode=simple&page=1')
soup = BeautifulSoup(page.content, 'html.parser')
movie_titles = soup.find_all(class_='titleColumn')
code_title_dict = {mov.a['href'].split('/')[2]:mov.a.text for mov in movie_titles}
```

```python
codes = []
for k in code_title_dict.keys():
    codes.append('&i=' + k)
codes
```

```python
api_key = '944239dd'
```

```python
def omdb_api_call(codes, api_key):
    new_list = []
    for imdbid in codes:
        url = 'http://www.omdbapi.com/?apikey=' + api_key + imdbid
        response = requests.get(url)
        results = json.loads(response.text)
        new_list.append(results)
    return new_list
```

```python
apikeycalling = omdb_api_call(codes, api_key)
apikeycalling
```

```python
def csv_saving(new_movie_dataframe,csv_filename):
    #temporary_df : bringing in blank csv file to update
    temporary_df = pd.read_csv(csv_filename, index_col =0)
    
    #current_df : adjusting current df with new attributes (etc...)
    current_df = pd.DataFrame(new_movie_dataframe, columns = columns)
    
    #temp_combined_df : temporary df placeholder with both current and previous df mixed together
    temp_combined_df = [temporary_df,current_df]
    
    #combined_df : final merged df
    combined_df = pd.concat(temp_combined_df).reset_index(drop=True) 
    
    #reuploading to csv file
    combined_df.to_csv('movie_rating_list.csv')
```

```python
csv_saving(new_movie_dataframe,'movie_rating_list.csv')
```
